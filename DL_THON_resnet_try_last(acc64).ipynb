```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np
from PIL import Image
import glob
import os
from sklearn.model_selection import train_test_split
```

```python
#바뀐 jellyfishes 의 데이터300개를 x_train,y_train데이터에 (학습데이터)에 할당해준다.

def load_data(img_path, number_of_data=900):  # 가위바위보 이미지 개수 총합에 주의하세요.
    img_size=224
    color=3
    #이미지 데이터와 라벨(Moon_jellyfish : 0, mauve_stinger_jellyfish : 1, lions_mane_jellyfish : 2 , compass_jellyfish:3 ,blue_jellyfish:4,barrel_jellyfish:5 ) 데이터를 담을 행렬(matrix) 영역을 생성합니다.
    imgs=np.zeros(number_of_data*img_size*img_size*color,dtype=np.int32).reshape(number_of_data,img_size,img_size,color)
    labels=np.zeros(number_of_data,dtype=np.int32)

    idx=0
    for file in glob.iglob(img_path+'/Moon_jellyfish/*'):#해당디렉토리에는 이미지파일만있고 확장자만 다르기때문에 전체를 불러옴
        img = np.array(Image.open(file),dtype=np.int32)
        imgs[idx,:,:,:]=img    # 데이터 영역에 이미지 행렬을 복사
        labels[idx]=0
        idx=idx+1

    for file in glob.iglob(img_path+'/mauve_stinger_jellyfish/*'):
        img = np.array(Image.open(file),dtype=np.int32)
        imgs[idx,:,:,:]=img    # 데이터 영역에 이미지 행렬을 복사
        labels[idx]=1   
        idx=idx+1  
    
    for file in glob.iglob(img_path+'/lions_mane_jellyfish/*'):
        img = np.array(Image.open(file),dtype=np.int32)
        imgs[idx,:,:,:]=img    # 데이터 영역에 이미지 행렬을 복사
        labels[idx]=2  
        idx=idx+1
        
    for file in glob.iglob(img_path+'/compass_jellyfish/*'):
        img = np.array(Image.open(file),dtype=np.int32)
        imgs[idx,:,:,:]=img    # 데이터 영역에 이미지 행렬을 복사
        labels[idx]=3
        idx=idx+1
        
    for file in glob.iglob(img_path+'/blue_jellyfish/*'):
        img = np.array(Image.open(file),dtype=np.int32)
        imgs[idx,:,:,:]=img    # 데이터 영역에 이미지 행렬을 복사
        labels[idx]=4
        idx=idx+1
        
    for file in glob.iglob(img_path+'/barrel_jellyfish/*'):
        img = np.array(Image.open(file),dtype=np.int32)
        imgs[idx,:,:,:]=img    # 데이터 영역에 이미지 행렬을 복사
        labels[idx]=5
        idx=idx+1
        
        
    print("학습데이터(x_train)의 이미지 개수는", idx,"입니다.")
    return imgs, labels
image_dir_path = os.path.join(os.getenv('HOME'), 'aiffel', 'jellyfish', 'Train_Test_Valid','Train')
#이미지데이터의 최솟값,최댓값을 확인후 0,1사이의 정규화를 통해 데이터를 정규화 시켜준다.
(x_train, y_train)=load_data(image_dir_path)#load_data함수로imgs, labels를 반환해서 각각 변수에할당
print('최소값:',np.min(x_train), ' 최대값:',np.max(x_train))
x_train_norm = x_train/255.0   # 입력은 0~1 사이의 값으로 정규화

print("x_train shape: {}".format(x_train.shape))
print("y_train shape: {}".format(y_train.shape))
```

```python
#진단할 resnet모델링
import tensorflow as tf
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, GlobalAveragePooling2D, Dense, BatchNormalization, ReLU, Add
from tensorflow.keras.initializers import HeNormal

def conv2d_bn(x, filters, kernel_size, strides, kernel_initializer, kernel_regularizer):
    x = Conv2D(filters, kernel_size, strides=strides, padding='same', 
               kernel_initializer=kernel_initializer,
               kernel_regularizer=kernel_regularizer)(x)
    x = BatchNormalization()(x)
    return x

def conv2d_bn_relu(x, filters, kernel_size, strides, kernel_initializer, kernel_regularizer):
    x = conv2d_bn(x, filters, kernel_size, strides, kernel_initializer, kernel_regularizer)
    x = Activation('relu')(x)
    return x
# def ResidualBlock(x, filters, kernel_size, padding, kernel_initializer, kernel_regularizer, downsample):
#     shortcut = x

#     if downsample:
#         # 1x1 합성곱을 사용하여 shortcut의 채널 수를 맞춰줍니다.
#         shortcut = Conv2D(filters, kernel_size=(1, 1), strides=(2, 2), padding='valid', kernel_initializer=kernel_initializer, kernel_regularizer=kernel_regularizer)(shortcut)
#         shortcut = BatchNormalization()(shortcut)

#     x = Conv2D(filters, kernel_size=kernel_size, strides=(1, 1), padding=padding, kernel_initializer=kernel_initializer, kernel_regularizer=kernel_regularizer)(x)
#     x = BatchNormalization()(x)
#     x = Activation('relu')(x)

#     x = Conv2D(filters, kernel_size=kernel_size, strides=(1, 1), padding=padding, kernel_initializer=kernel_initializer, kernel_regularizer=kernel_regularizer)(x)
#     x = BatchNormalization()(x)

#     # downsample이 False인 경우에도 채널 수가 다른 경우에 대비하여 1x1 합성곱을 추가합니다.
#     if shortcut.shape[-1] != x.shape[-1]:
#         shortcut = Conv2D(filters, kernel_size=(1, 1), strides=(1, 1), padding='valid', kernel_initializer=kernel_initializer, kernel_regularizer=kernel_regularizer)(shortcut)
#         shortcut = BatchNormalization()(shortcut)

#     out = Add()([shortcut, x])
#     out = Activation('relu')(out)
#     return out

# ResidualBlock 함수를 정의합니다.
def ResidualBlock(x, filters, kernel_size, padding, kernel_initializer, kernel_regularizer, downsample=False):
#(x, filters, kernel_size, kernel_initializer,  downsample=False):
    shortcut = x
    if downsample:
        shortcut = conv2d_bn(x, filters, kernel_size=1, strides=2, kernel_initializer=kernel_initializer, kernel_regularizer=kernel_regularizer)
        strides = 2
    else:
        strides = 1

#     x = conv2d_bn_relu(x, filters, kernel_size, strides=strides, kernel_initializer=kernel_initializer)
    x = conv2d_bn_relu(x, filters, kernel_size, strides=strides, kernel_initializer=kernel_initializer, kernel_regularizer=kernel_regularizer)
#     x = conv2d_bn(x, filters, kernel_size, strides=1, kernel_initializer=kernel_initializer)
    x = conv2d_bn(x, filters, kernel_size , strides=1, kernel_initializer=kernel_initializer, kernel_regularizer=kernel_regularizer)

    # Add() 레이어를 사용하여 shortcut connection을 구현합니다.
    out = Add()([shortcut, x])
    out = Activation('relu')(out)
    return out
```

```python
#resnet모델만들기
import tensorflow as tf
from tensorflow.keras.layers import Input, AveragePooling2D,Flatten,Activation,Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.regularizers import l2
def ResNet18(classes, input_shape):
    inputs = Input(shape=input_shape)
    conv1 = Conv2D(32, (7,7), strides=2, padding='valid', activation='relu', kernel_initializer='he_normal')(inputs)
#     conv1 = Conv2D(32, (5,5), strides=2, padding='same', activation='relu', kernel_initializer='he_normal', kernel_regularizer=l2(0.01))(inputs)
    
    max_pooling = MaxPooling2D(3, strides=2, padding='valid')(conv1)
   # max_pooling = MaxPooling2D(3, strides=2)(conv1)
    conv2_x = ResidualBlock(max_pooling, 32, kernel_size=(3,3), padding='same', kernel_initializer='he_normal', kernel_regularizer=l2(0.1), downsample=True)
    conv2_x = ResidualBlock(max_pooling, 32, kernel_size=(3,3),  padding='same',kernel_initializer='he_normal', kernel_regularizer=l2(0.1), downsample=False)

#     conv2_x = ResidualBlock(max_pooling, 64, kernel_size=(3,3), kernel_initializer='he_normal', downsample=False)
#     conv2_x = ResidualBlock(conv2_x, 64, kernel_size=(3,3), kernel_initializer='he_normal',  downsample=False)
    
    conv3_x = ResidualBlock(conv2_x, 64,kernel_size=(5,5), padding='same', kernel_initializer='he_normal', kernel_regularizer=l2(0.3), downsample=True)
    conv3_x = ResidualBlock(conv3_x, 64,kernel_size=(5,5),  padding='same',kernel_initializer='he_normal', kernel_regularizer=l2(0.3), downsample=False)

#     conv3_x = ResidualBlock(conv2_x, 128, kernel_size=(3,3), kernel_initializer='he_normal',  downsample=True)
#     conv3_x = ResidualBlock(conv3_x, 128, kernel_size=(3,3), kernel_initializer='he_normal', downsample=False)
    
    conv4_x = ResidualBlock(conv3_x, 64, kernel_size=(5,5),  padding='same',kernel_initializer='he_normal', kernel_regularizer=l2(0.1), downsample=True)
    conv4_x = ResidualBlock(conv4_x, 64, kernel_size=(5,5),  padding='same',kernel_initializer='he_normal', kernel_regularizer=l2(0.1), downsample=False)

#     conv4_x = ResidualBlock(conv3_x, 256, kernel_size=(3,3), kernel_initializer='he_normal', downsample=True)
#     conv4_x = ResidualBlock(conv4_x, 256, kernel_size=(3,3), kernel_initializer='he_normal',  downsample=False)
    
    conv5_x = ResidualBlock(conv4_x, 64, kernel_size=(5,5),  padding='same',kernel_initializer='he_normal', kernel_regularizer=l2(0.1), downsample=True)
    conv5_x = ResidualBlock(conv5_x, 64, kernel_size=(5,5),  padding='same',kernel_initializer='he_normal', kernel_regularizer=l2(0.1), downsample=False)

#     conv5_x = ResidualBlock(conv4_x, 512, kernel_size=(3,3), kernel_initializer='he_normal', downsample=True)
#     conv5_x = ResidualBlock(conv5_x, 512, kernel_size=(3,3), kernel_initializer='he_normal',  downsample=False)
    
#     x = GlobalAveragePooling2D()(conv5_x)
#     x = Flatten()(x)
    x = Flatten()(conv5_x)
    x = Dense(32, activation='relu', kernel_initializer='he_normal', kernel_regularizer=l2(0.1))(x)
    x = Dropout(0.3)(x)
    x = Dense(32, activation='relu', kernel_initializer='he_normal', kernel_regularizer=l2(0.1))(x)
#     x = Dropout(0.5)(x)
    out_layer = Dense(units=classes, activation='softmax')(x)
    
    model = Model(inputs=inputs, outputs=out_layer)
    return model
# 커널사이즈, DENSE층노드수
```
